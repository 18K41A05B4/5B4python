```python
#5. write a NumPy program to compute the sum of the diagonal element of a given array(matrix).
import numpy as np
m=np.arange(9).reshape(3,3)
print("original matrix:")
print(m)
r=np.trace(m)
print("sum of diagonal element:",r)
```

```python
4.#Write a numpy program to compute the condition number of a given matrix.
import numpy as np
s=np.array([[5,6],[3,4]])
print("matrix:")
print(s)
c=np.linalg.cond(s)
print("The condition number of a given matrix:",c)
```
